```python
#Import all relevant libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
warnings.filterwarnings("ignore", category=DeprecationWarning) 
```

```python
#Load the file into pandas 
df = pd.read_csv("/Users/aromerq/Documents/Exploración/market_basket/trx_data_80_h.csv")
df.head()
```

```python
#Check the structure of the dataframe
df.info()
```

```python
#dimensions dataset
df.shape
```

```python
#count null values
df.isna().sum()
```

```python
#Drop all rows with a null value
df.dropna(inplace=True)
len(df)
```

```python
#Convert the trx_id column to string
df["trx_id"] = df["trx_id"].astype('str')

len(df)
```

```python
#Group, sum, unstack and set index of dataframe
basket = df.groupby(['trx_id', 'Subcategoria'])["sales_unit_qty"]\
        .sum().unstack()\
        .reset_index().fillna(0)\
        .set_index("trx_id")

basket.head()
```

```python
#Create function to hot encode the values
def encode_values(x):
    if x <= 0:
        return 0
    if x > 0:
        return 1

#Apply function to data
basket_encoded = basket.applymap(encode_values)

basket_encoded
```

```python
#filter for only trx with 2 or more items
basket_filtered = basket_encoded[(basket_encoded > 0).sum(axis=1) >= 1]

basket_filtered
```

```python
#Fill na to null
basket_filtered_ = basket_filtered.fillna(0)
basket_filtered_
```

```python
#basket_filtered_.to_csv("basket_filtered_.csv")
```

```python
#Generate y apply to algoritm apriori
frequent_itemsets = apriori(basket_filtered_, min_support=0.03, use_colnames=True).sort_values("support",ascending=False).reset_index(drop=True)
frequent_itemsets.head(30)

```

```python
data1 = frequent_itemsets.loc[frequent_itemsets['itemsets'].apply(lambda conjunto: conjunto.issuperset({'23284  ASADURIAS'}))]
data1
```

```python
#frequent_itemsets.to_csv("frequent_itemsets_80.csv")
```

```python
#Generate y apply algoritm association rules
assoc_rules = association_rules(frequent_itemsets, metric="lift", min_threshold=0.7).sort_values("lift",ascending=False).reset_index(drop=True)
```
